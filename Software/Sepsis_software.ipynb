```python
import random
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# this function generates a random set of inputs, resembling a potentiostat input for each of the biomarkers
def get_rand():
    return random.uniform(0,1000)
```

```python
rn = 10 # number of random inputs we have
```

```python
#making dataframes to store all outputs
il1b_plasma_list = []
il_6_list = []
tnf_a_list = []
crp_list = []
lactoferrin_list = []
```

```python
# this part generates the output, i.e the plasma value for the 

for i in range(0, rn):
    
    # here we can change the equations to fit for any disease/
    # biomarker we need as long as we have modelled equations
    # each line is formatted as follows:
    # biomarker = a + b * (potentiostat reading)
    # more lines can be added for more biomarkers or change the existing ones
    il1b_plasma = -0.859121099839232 + 0.9114676655388801*(float(get_rand()))
    il_6 = 1.5704654392436872 + 0.6839493493025036*(float(get_rand()))
    tnf_a = -0.3610625115091586 + 0.7903607861416451*(float(get_rand()))
    crp = float(get_rand())
    lactoferrin = float(get_rand())
    
    # storing values in a list
    il1b_plasma_list.append(il1b_plasma)
    il_6_list.append(il_6)
    tnf_a_list.append(tnf_a)
    crp_list.append(crp)
    lactoferrin_list.append(lactoferrin)
    #print(il1b_plasma_list,il_6_list, tnf_a_list, crp_list, lactoferrin_list)
```

```python
#PLOTTING ALL VARIABLES
plot1 = plt.plot(il1b_plasma_list, label='IL1B')
plot2 = plt.plot(il_6_list, label='IL6')
plot3 = plt.plot(tnf_a_list, label='TNFA')
plot4 = plt.plot(crp_list,label='CRP')
plot5 = plt.plot(lactoferrin_list,label='LACTOFERRIN')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
```

```python
# Add disease name here
disease_name = "Sepsis"
```

```python
# checking each value against a threshold, a true output means that there is a 
# value above the threshold received from modelling
# a value of true determines if there is sepsis risk for 
# the value of the biomarker in the dataset 
for val in il1b_plasma_list:
    if val >= 0.33:
        print(f" {disease_name} risk exists for {val} in il1b")
for val in il_6_list:
    if val >= 16:
        print(f"{disease_name} risk exists for {val} in lil6")
for val in tnf_a_list:
    if val >= 5:
        print(f"{disease_name} risk exists for {val} in tnfa")
for val in crp_list:
    if val >= 80:
        print(f"{disease_name} risk exists for {val} in crp")
for val in lactoferrin_list:
    if val >= 675:
        print(f"{disease_name} risk exists for {val} in lactoferrin")
```
